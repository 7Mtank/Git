```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have a DataFrame df with columns 'area', 'bedrooms', 'age', and 'price'
# Replace 'your_dataset.csv' with your actual dataset file if needed
# df = pd.read_csv('your_dataset.csv')

# Extract the variables
area = df['area']
bedrooms = df['bedrooms']
age = df['age']
price = df['price']

# Create a scatter plot for each variable against the target variable (price)
plt.figure(figsize=(12, 6))

plt.subplot(131)  # Subplot for area vs. price
plt.scatter(area, price, alpha=0.5)
plt.title('Area vs. Price')
plt.xlabel('Area')
plt.ylabel('Price')


plt.subplot(132)  # Subplot for bedrooms vs. price
plt.scatter(bedrooms, price, alpha=0.5)
plt.title('Bedrooms vs. Price')
plt.xlabel('Bedrooms')
plt.ylabel('Price')

plt.subplot(133)  # Subplot for age vs. price
plt.scatter(age, price, alpha=0.5)
plt.title('Age vs. Price')
plt.xlabel('Age')
plt.ylabel('Price')

plt.tight_layout()  # To ensure proper spacing between subplots

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
#area = [1200, 1400, 1600, 1800, 2000]
#bedrooms = [2, 3, 3, 4, 4]
#age = [10, 5, 15, 10, 5]
#price = [220000, 250000, 280000, 300000, 320000]

# Create a scatter plot
plt.figure(figsize=(12, 6))

# Scatter plot of price vs. area
plt.subplot(131)
plt.scatter(area, price, marker='o', c='b', label='Price vs. Area')
plt.xlabel('Area (sq. ft)')
plt.ylabel('Price ($)')
plt.grid(True)

# Scatter plot of price vs. bedrooms
plt.subplot(132)
plt.scatter(bedrooms, price, marker='o', c='g', label='Price vs. Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Price ($)')
plt.grid(True)

# Scatter plot of price vs. age
plt.subplot(133)
plt.scatter(age, price, marker='o', c='r', label='Price vs. Age')
plt.xlabel('Age (years)')
plt.ylabel('Price ($)')
plt.grid(True)

# Fit a linear regression line for price vs. area
coeff_area = np.polyfit(area, price, 1)
fit_area = np.poly1d(coeff_area)
plt.subplot(131)
plt.plot(area, fit_area(area), '--k', label='Best-fit Line')

# Fit a linear regression line for price vs. bedrooms
coeff_bedrooms = np.polyfit(bedrooms, price, 1)
fit_bedrooms = np.poly1d(coeff_bedrooms)
plt.subplot(132)
plt.plot(bedrooms, fit_bedrooms(bedrooms), '--k', label='Best-fit Line')

# Fit a linear regression line for price vs. age
coeff_age = np.polyfit(age, price, 1)
fit_age = np.poly1d(coeff_age)
plt.subplot(133)
plt.plot(age, fit_age(age), '--k', label='Best-fit Line')

# Add legends
plt.subplot(131)
plt.legend()
plt.subplot(132)
plt.legend()
plt.subplot(133)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
